## Observations and Insights 

Based on the data collected for the treatment, we can conclude three oberservations as following:

1. The summary statistics table, bar and pie charts indicated all treatments were done very similarly in regards to the average tumor volumes, number of mice undertook different treatments and gender of mice to avoid bias and ambigious result. It is very intersting to see that keep those factors around the same in order to get the consistency for the treatment results. 

2. From the Quartiles, Outliers and Boxplots section, it was easier to compare the final tumor volume of each mouse across four regimens and the company's interest -Capomulin is performing better compared to the other treatment regimens' tumor volume. Visually, the boxplot for Capomulin is smaller compared to other treatment regimens and has no outlier. 

3. From the Line and Scatter Plots section, it looks at the specific treatment on a random selected mouse. The tumor volume is gradually decreased based on time spent on for the Capomulin treatment. The line is gradually decreasing which can be easily noticed. On the other hand, the scatter plots section has measured the relationships between the tumor volume compared to the weight of mice. We can notice that the higher mice weight, they are more likely to have a large size of tumor volume from using the Capomulin. 

To conclude based on the data collected and analysed, Capomulin is showing great performance compared to other drug regimens and the longer treatment used, the smaller tumor volume became. However, it is also testified that the treatment won't be very effective on high weight patients.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
# Merge functions on Mouse ID 
combined_df = pd.merge(mouse_metadata, study_results, on ="Mouse ID", how="outer")

# Display the data table for preview
combined_df.head()
```

```python
# identify incomplete rows

combined_df.count()
```

```python
# Checking the number of mice.
mice=combined_df["Mouse ID"].value_counts()

number_of_mice=len(mice)

number_of_mice

```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate = combined_df[combined_df.duplicated(["Mouse ID", "Timepoint"])]
print("Duplicate mice based on 2 columns are:", duplicate , sep='\n')
```

```python

# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_df.loc[combined_df.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
duplicate_mice 
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  

# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint


Capomulin_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:]

Capomulin_last = Capomulin_df.groupby('Mouse ID').max()['Timepoint']
Capomulin_vol = pd.DataFrame(Capomulin_last)
Capomulin_merge = pd.merge(Capomulin_vol, clean_df, on=("Mouse ID","Timepoint"),how="left")
Capomulin_merge.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
Capomulin_tumors = Capomulin_merge["Tumor Volume (mm3)"]

quartiles = Capomulin_tumors.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of Capomulin tumors is: {lowerq}")
print(f"The upper quartile of Capomulin tumors is: {upperq}")
print(f"The interquartile range of Capomulin tumors is: {iqr}")
print(f"The the median of Capomulin tumors is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
average_cap = Capomulin_df.groupby('Mouse ID').mean()
x_values = average_cap["Weight (g)"]
y_values = average_cap["Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,36),fontsize=15,color="red")
                               

plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title("Mouse weight versus average tumor volume for the Capomulin regimen")
plt.show()
print(f"The r-squared is: {rvalue**2}")
```
